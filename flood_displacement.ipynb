```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle

# Setting Seaborn theme
sns.set_style("whitegrid")

# Using the 'deep' color palette
colors = sns.color_palette("deep")

# Data
categories = {
    'Education': {
        'labels': ['No schooling', 'up to Matric', 'HSSC and above'],
        'sizes': [32.82, 51.28, 15.90]
    },
    'Family size': {
        'labels': ['Small (up to 4 people)', 'Medium (5-6 people)', 'Above 7 people'],
        'sizes': [9.74, 20, 70.26]
    },
    'Personal Vehicle ownership': {
        'labels': ['Car, Tractor Truck', 'Donkey cart', 'Motorbike', 'None'],
        'sizes': [34.36, 43.59, 75.90, 8.21]
    },
    'House ownership': {
        'labels': ['Owner', 'Rented', 'Mixed'],
        'sizes': [80.51, 10.26, 9.23]
    },
    'House type': {
        'labels': ['Cemented', 'Mud', 'Mixed'],
        'sizes': [30.77, 35.90, 33.33]
    },
    'House damage': {
        'labels': ['No Damage', '1/4th Damage', 'Half Damage', '3/4th Damage', 'Totally Damage'],
        'sizes': [19.49, 27.69, 38.46, 11.28, 3.08]
    }
}

# Create subplots
fig, axs = plt.subplots(2, 3, figsize=(20, 14), facecolor='white')

# Plotting function
def plot_data(key, ax):
    data = categories[key]
    wedges, texts, autotexts = ax.pie(
        data['sizes'], 
        labels=data['labels'], 
        autopct='%1.2f%%', 
        colors=colors[:len(data['labels'])], 
        radius=1.2,
        wedgeprops=dict(width=0.3, edgecolor='w')
    )

    # Setting text elements
    ax.set_title(key, fontsize=24)
    plt.setp(texts, size=20)  # Adjusting label size
    plt.setp(autotexts, size=18)  # Adjusting percentage size
    ax.set_facecolor('white')  # Setting the pie chart background to white

    # Adding a rectangle around the pie chart
    box = Rectangle((-1.3, -1.3), 2.6, 2.6, fill=False, edgecolor='black', linewidth=1.5)
    ax.add_patch(box)

# Plotting data in the specified order
plot_data('Education', axs[0, 0])
plot_data('Family size', axs[0, 1])
plot_data('Personal Vehicle ownership', axs[0, 2])
plot_data('House ownership', axs[1, 0])
plot_data('House type', axs[1, 1])
plot_data('House damage', axs[1, 2])

# Adjust layout to decrease spaces between plots
plt.subplots_adjust(wspace=0.4, hspace=0.2)

# Set the main title for the whole plot with normal weight
fig.suptitle("Demographic details of Respondents.", fontsize=26, y=0.95)

# Save the figure as a PNG with 300 dpi
plt.savefig("demo_data.png", dpi=300, bbox_inches='tight', facecolor='white')

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('Path')


# Define colors using the "deep" color palette
colors = sns.color_palette("deep", 8)

# Set seaborn styles
sns.set_style("whitegrid")
sns.set_palette(colors)

# Compute the correlation matrix
corr = data.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 6))
ax = sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".2f",
                 mask=mask,  # Apply the mask
                 cbar_kws={'label': 'Correlation coefficient'},
                 annot_kws={"size": 12, "color": "black"})

# Setting the title and axis labels with specific font sizes
plt.title('Correlation Heatmap', fontsize=18)

# Increase Font size for x and y axis labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Rotate x and y axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, horizontalalignment='right')
ax.set_yticklabels(ax.get_yticklabels(), rotation=00, verticalalignment='center')

# Increase font size of color bar labels
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=12)

# Increase font size of the vertical color bar title
cbar.ax.yaxis.label.set_size(16)

# Save the figure
plt.savefig("C:/Users/u301380/Documents/pyhton2022/NCA_paper/heatmapfinal.png", dpi=300, bbox_inches='tight', facecolor='white')

# Show the plot (optional)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting seaborn style
sns.set_style("whitegrid")

# Modified dataframe with your provided data
data = {
    'Construct': ['Fear', 'Response efficacy', 'Severity', 'Protective cost', 'Vulnerability', 'Self-efficacy', 'Reward'],
    'Importance': [0.489, 0.324, 0.100, 0.057, 0.036, 0.024, 0.003],
    'Performance': [77.55, 53.809, 89.133, 67.444, 80.66, 35.432, 72.381],
    'Rank_in_Importance': [1, 2, 3, 4, 5, 6, 7],
    'Rank_in_Performance': [3, 6, 1, 5, 2, 7, 4]
}
df = pd.DataFrame(data)

# Plotting with size adjustments
plt.figure(figsize=(10, 8))

# Adjust the size more drastically based on Rank_in_Importance
sizes = df['Rank_in_Importance'].apply(lambda x: (8-x)*200)

# Adjust color intensity based on Rank_in_Performance with a more distinguishable palette
color_palette = sns.color_palette("deep", 7)
colors = df['Rank_in_Performance'].map(lambda rank: color_palette[rank-1])

# Scatter plot and annotations
for i, row in df.iterrows():
    plt.scatter(row['Importance'], row['Performance'], color=colors[i], s=sizes[i], label=row['Construct'])
    plt.annotate(row['Construct'], (row['Importance'], row['Performance']), fontsize=16, ha='right')

plt.xlabel('Importance', fontsize=18)
plt.ylabel('Performance', fontsize=18)
plt.title('Importance-Performance Map Analysis (IPMA) of Predictors', fontsize=18)
plt.grid(True)

# Modified legend that indicates the construct and rank in performance
handles, labels = plt.gca().get_legend_handles_labels()
sorted_handles = [x for _, x in sorted(zip(df['Rank_in_Performance'], handles))]
sorted_labels = [x for _, x in sorted(zip(df['Rank_in_Performance'], labels))]

# Axes adjustments
plt.xlim(-0.1, 0.50)
plt.ylim(1, 100)
plt.xticks(fontsize=18)
plt.yticks(range(10, 101, 10), fontsize=18)

plt.tight_layout()
plt.savefig("C:/Users/u301380/Documents/pyhton2022/NCA_paper/imp21.png", dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
